# Using GAM on Abalone dataset

```python
# load data
from pandas import read_csv

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data'
names = [
    'sex', 
    'length', 
    'diameter', 
    'height', 
    'weight_whole', 
    'weight_shucked', 
    'weight_viscera', 
    'weight_shell',
    'rings']
dataframe = read_csv(url, names=names, index_col=False)

dataframe.head(2)
```

```python
#Create some 'fake', random features.
import numpy as np

dataframe['f1'] = np.random.randint(1, 100, dataframe.shape[0])
dataframe['f2'] = np.random.randint(1, 100, dataframe.shape[0])
dataframe['f3'] = np.random.randint(1, 100, dataframe.shape[0])
dataframe['f4'] = np.random.randint(1, 100, dataframe.shape[0])
dataframe['f5'] = np.random.randint(1, 100, dataframe.shape[0])

dataframe.head(2)
```

```python
# note I'm excluding sex, as I don't want to bother with categorical for now
X = dataframe.values[:, range(1,8) + range(9,dataframe.shape[1])]
y = dataframe.values[:, 8]

print(X.shape)
print(y.shape)
```

## Plot

I want to see how the data looks against y

```python
import matplotlib.pyplot as plt
```

```python
cols = 3

if X.shape[1] % cols == 0: 
    rows = X.shape[1] // cols
else:
    rows = X.shape[1] // cols + 1
    
dims = [rows, cols]

fig, axs = plt.subplots(nrows=dims[0], ncols=dims[1], figsize=(17,10))
plt.tight_layout()

for row in range(dims[0]):
    for col in range(dims[1]):
        index = row*dims[1] + col
        
        if index < X.shape[1]:
            tmp = X[:,index]
            axs[row, col].scatter(tmp,y)
            axs[row, col].set_title(dataframe.columns[index+1])
plt.show()
```

# Simple GAM

```python
from pygam import LinearGAM, s, l, f

for i in range(X.shape[1]):
    if i == 0:
        terms = s(i, n_splines=5)
    else:
        terms = terms + s(i, n_splines=5)

terms

gam = LinearGAM(terms).fit(X, y)
gam.summary()
```

```python
fig, axs = plt.subplots(nrows=dims[0], ncols=dims[1], figsize=(17,10))
plt.tight_layout()

for row in range(dims[0]):
    for col in range(dims[1]):
        i = row*dims[1] + col
        
        if i < len(gam.terms)-1:
            term = gam.terms[i]
        
            if term.isintercept:
                continue
        
            XX = gam.generate_X_grid(term=i)
            pdep, confi = gam.partial_dependence(term=i, X=XX, width=0.95)
            axs[row, col].plot(XX[:, term.feature], pdep)
            axs[row, col].plot(XX[:, term.feature], confi, c='r', ls='--')
            axs[row, col].set_title(str(repr(gam.terms[i])))

plt.show()
```

# Feature selector - by 'parsimony'

```python
def make_terms(index):
    for i in index:
        if i == 0:
            terms = s(i, n_splines=5)
        else:
            terms = terms + s(i, n_splines=5)
    return terms

make_terms(range(5))
```

```python
features_index_all = range(12)
features_index = features_index_all
is_parsim = False
```

```python
import operator
```

```python


while is_parsim == False:
    print("Running")
    
    # make model terms
    terms_tmp = make_terms(features_index)
    # estimate model
    gam_tmp = LinearGAM(terms_tmp).fit(X, y)
    
    p_vals = gam_tmp.statistics_['p_values'][0:len(gam_tmp.statistics_['p_values'])-1]
    
    if max(p_vals) > .05 :
        index, value = max(enumerate(p_vals), key=operator.itemgetter(1))
        print(index)
        del features_index[index]
        
    else:
        is_parsim = True

print("final features selected")
print(features_index)
```

```python
gam_tmp.summary()
```

```python
fig, axs = plt.subplots(nrows=dims[0], ncols=dims[1], figsize=(17,10))
plt.tight_layout()

for row in range(dims[0]):
    for col in range(dims[1]):
        i = row*dims[1] + col
        
        if i < len(gam_tmp.terms)-1:
            term = gam_tmp.terms[i]
        
            if term.isintercept:
                continue
        
            XX = gam_tmp.generate_X_grid(term=i)
            pdep, confi = gam.partial_dependence(term=i, X=XX, width=0.95)
            axs[row, col].plot(XX[:, term.feature], pdep)
            axs[row, col].plot(XX[:, term.feature], confi, c='r', ls='--')
            axs[row, col].set_title(str(repr(gam_tmp.terms[i])))

plt.show()
```
